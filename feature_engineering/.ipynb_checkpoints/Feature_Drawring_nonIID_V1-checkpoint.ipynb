```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import random
import os
from tqdm import tqdm
from sklearn.model_selection import train_test_split
```

```python
def read_all_csv_files(directory_path, features, limit=1000000):
    # Initialize an empty list to store dataframes from CSV files
    dataframes = []

    # Get a list of all files in the directory
    file_list = os.listdir(directory_path)

    # Loop through each file and check if it's a CSV file
    for file_name in file_list:
        if file_name.endswith('.csv'):
            # Get the full file path
            file_path = os.path.join(directory_path, file_name)
            # Read the CSV file into a pandas DataFrame
            df = pd.read_csv(file_path)
            # Remove leading and trailing spaces from column names
            df.columns = df.columns.str.strip()
            df['Stratify'] = file_name
            # Append the DataFrame to the list
            dataframes.append(df[features])

    # Merge all DataFrames into a single DataFrame
    merged_df = pd.concat(dataframes, ignore_index=True)
    print("Initial Row: {}".format(merged_df.shape[0]))
    
    #applying limit if row_count > limit
    if len(merged_df) > limit:
        print("As Row Exceeds {0}, we are taking upto {1} rows".format(limit, limit))
        #merged_df = merged_df.sample(limit, random_state=42)
        sampled_df, _ = train_test_split(
            merged_df, 
            train_size=limit,  # Select 'limit' rows
            stratify=merged_df['Stratify'], 
            random_state=42
        )

        return sampled_df
    
    return merged_df
```

```python
def sanitize_data_frames_updated(dataframe, limit=1000000, remove_infinity=True, remove_null=True):
    print("Before Santization Row Count:", dataframe.shape[0])

    if remove_infinity:
        numeric_cols = dataframe.select_dtypes(include=[np.number]).columns
        infinite_counts = dataframe[numeric_cols].applymap(np.isinf).sum()
        for col, count in infinite_counts.items():
            if count != 0:
                print(f"Column '{col}' contains {count} infinite values.")
                if limit is not None:
                    dataframe = dataframe[~np.isinf(dataframe[col])]

    if remove_null:
        null_counts = dataframe.isnull().sum()
        for col, count in null_counts.items():
            if count != 0:
                print(f"Column '{col}' contains {count} null values.")
                if limit is not None:
                    dataframe = dataframe.dropna(subset=[col])
    
    print("Sanitized Row Count:", dataframe.shape[0])    
    return dataframe
```

```python
##Function to draw the box plot
def draw_box_plot(data, tick_list, title, x_label="", y_label="", show_outliers=False, vertical=True, size=(6,3), tick_color=[]):
    #Finding the Tick Numbers

     # Make sure the length of tick_color list matches the length of tick_list
    if tick_color and len(tick_color) != len(tick_list):
        raise ValueError("tick_color list must match the length of tick_list")
    
    tick_numbers = [index+1 for index, value in enumerate(tick_list)]
    fig, ax = plt.subplots(figsize=size)
    plt.suptitle(title)
    ax = fig.add_axes([0, 0, 1, 1])
    bp = ax.boxplot(data, showfliers=show_outliers, vert=vertical, patch_artist=True)
    #bp = ax.boxplot(data, showfliers=show_outliers, vert=vertical)
    plt.xticks(tick_numbers, tick_list, rotation=90)    
    #plt.xticks(rotation=90)

    # Set the color of the tick labels if tick_color list is provided
    if len(tick_color) > 0:
        for tick_label, color in zip(ax.get_xticklabels(), tick_color):
            tick_label.set_color(color)
    
    ax.set_ylabel(y_label)
    # plt.yscale('log')
    plt.show()
```

```python
##Function to draw the box plot
def draw_box_plot_with_outliers_old(data, tick_list, title, x_label="", y_label="", show_outliers=False, vertical=True):
    data = np.array(data)
    # Calculate outliers for each category
    outliers = [identify_outliers(category) for category in data]
    
    #Finding the Tick Numbers
    tick_numbers = [index+1 for index, value in enumerate(tick_list)]
    fig, ax = plt.subplots(figsize=(6,3))    
    plt.suptitle(title)
    ax = fig.add_axes([0, 0, 1, 1])
    
    bp = ax.boxplot(data, sym='', widths=0.6)

    # Overlay the outliers
    for i, category_outliers in enumerate(outliers):
        # Scatter plot for outliers
        y = category_outliers
        x = np.random.normal(1+i, 0.04, size=len(y))
        ax.plot(x, y, 'r+', alpha=0.5)
    
    plt.xticks(tick_numbers, tick_list)    
    plt.xticks(rotation=90)
    ax.set_ylabel(y_label)
    # plt.yscale('log')
    # Setting grid
    ax.grid(True)
    plt.show()
```

```python
##Function to draw the box plot
def draw_box_plot_with_outliersssss(data_lists, tick_list, title, x_label="", y_label="", show_outliers=False, vertical=True):
    #print(len(data_lists))
    fig, ax = plt.subplots(figsize=(14,8))
    positions = np.arange(1, len(data_lists) + 1)

    # Plot boxplot for each category
    for position, data in zip(positions, data_lists):
        # Convert category data to numpy array for processing
        data_np = np.array(data)
        ax.boxplot(data_np, positions=[position], widths=0.6)
        
        # Identify and overlay outliers
        outlier_mask = identify_outliers(data_np)
        # Use the mask to select outliers
        outliers = data_np[outlier_mask]
        #print(outliers)
        # Generate random x positions for each outlier for plotting
        x_outliers = position + 0.1 * np.random.randn(len(outliers))
        #print(x_outliers)
        ax.plot(x_outliers, outliers, 'r+', alpha=0.05)
        
    #Finding the Tick Numbers
    tick_numbers = [index+1 for index, value in enumerate(tick_list)]
      
    plt.suptitle(title)
    ax = fig.add_axes([0, 0, 1, 1]) 
   
    plt.xticks(tick_numbers, tick_list)    
    plt.xticks(rotation=90)
    ax.set_ylabel(y_label)
    # plt.yscale('log')
    # Setting grid
    ax.grid(True)
    plt.show()
```

```python
def plot_with_outliers(data):
    fig, ax = plt.subplots(figsize=(14, 8))

    # Process each list in the data separately
    for i, category in enumerate(data):
        if isinstance(category, list) and len(category) > 0:
            # Convert the category to a numpy array
            category_array = np.array(category)
            
            # Identify outliers
            outlier_mask = identify_outliers(category_array)
            outliers = category_array[outlier_mask]

            # Plot the box for this category
            ax.boxplot(category_array, positions=[i+1], widths=0.6, sym='')

            # Overlay the outliers
            x_outliers = np.random.normal(i+1, 0.04, size=len(outliers))
            ax.plot(x_outliers, outliers, 'r+', alpha=0.5)

    # Set the ticks and labels for x-axis
    ax.set_xticks(range(1, len(data) + 1))
    ax.set_xticklabels(['Category {}'.format(i+1) for i in range(len(data))])
    ax.set_ylabel('Values')
    ax.set_title('Box Plot with Overlayed Outlier Points')
    ax.grid(True)
    plt.show()




```

```python
# Function to identify outliers within each list
def identify_outliers(data):
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1
    lower_fence = q1 - 1.5 * iqr
    upper_fence = q3 + 1.5 * iqr
    return (data < lower_fence) | (data > upper_fence)
```
